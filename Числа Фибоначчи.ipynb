```python
from rcviz import viz, CallGraph
from functools import lru_cache
```

```python
cg = CallGraph(filename="factorial.png")

cache = {}

@viz(cg)
def fib_memo(n):
    assert n >= 0
    if n not in cache:
        cache[n] = n if n <= 1 else fib_memo(n - 1) + fib_memo(n-2)
    return cache[n]

fib_memo(5)
cg.render()
```

```python
@lru_cache(maxsize=None) 
def fib_bad(n):
    assert n >= 0
    return n if n <= 1 else fib_bad(n - 1) + fib_bad(n - 2) 

fib_bad(1480)
```

```python
def fib(n):
    f0, f1 = 0, 1
    for _ in range(n - 1):
        f0, f1 = f1, f0 + f1
    return f1

def main():
    n = int(input())
    print(fib(n))


if __name__ == "__main__":
    main()
```

```python
import numpy as np 


def dot(vec_1, vec_2):
    return sum([i * j for i, j in zip(vec_1, vec_2)])


def matrix_multiply(matrix_1, matrix_2, m):
    result = np.zeros((len(matrix_1[0]), len(matrix_2)))
    for i in range(len(matrix_1)):
        for j in range(len(matrix_2[0])):
            result[i][j] = dot(matrix_1[i, :], matrix_2[:, j])
    return result % m 


def pow_matrix(matrix, n, m):
    if n == 0:
        return np.array(
            [[1, 0],
             [0, 1]]) 
    if n % 2 == 1:
        return matrix_multiply(pow_matrix(matrix, n-1, m), matrix, m)
    else:
        matrix = pow_matrix(matrix, n // 2, m)
        return matrix_multiply(matrix, matrix, m)
    
    
def fib_mod(n, m):
    fib_matrix = np.array(
        [[1, 1],
         [1, 0]])
    return pow_matrix(fib_matrix, n-1, m)[0][0]


def main():
    n, m = map(int, input().split())
    print(int(fib_mod(n, m)))


if __name__ == "__main__":
    main()
```

```python
def fib_mod(n, m):
    F=[0,1]
    c = 0
    for i in range(2, m * 6):
        F.append((F[i-1] + F[i-2]) % m)
        c += 1
        if F[i-1] == 0 and F[i] == 1:
            break
    return F[(n % (len(F) - 2))]


def main():
    n, m = map(int, input().split())
    print(int(fib_mod(n, m)))


if __name__ == "__main__":
    main()
```
